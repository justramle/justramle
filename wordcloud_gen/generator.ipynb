### Plotting word in wordcloud
- Example: https://stackoverflow.com/questions/65749888/how-to-generate-word-cloud-with-symbol
- Use regex pattern: any upper-lowecase  and + sign for C++ OR boundary with whole word include characters
    - regexp=r'[a-zA-Z\+]+|\b\w+\b'


```python
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
from PIL import Image
import random

def pastel_color(word, font_size, position, orientation, random_state=None, **kwargs):
    return "hsl(%d, %d%%, %d%%)" % (np.random.randint(220, 313), np.random.randint(30, 50), np.random.randint(65, 80))

def plot_word_cloud():
    # Read file from txt
    with open ('worddoc.txt', 'r', encoding='UTF-8') as file:
        wordlist = ' '.join([line.strip() for line in file])
    
    # Create cat shape mask
    mask = np.array(Image.open('../images/heart.png'))
    
    # Generate the word cloud 
    wordcloud = WordCloud(width=400, height=500, 
                          background_color='white',
                        #   stopwords= STOPWORDS,
                          font_path='Montserrat-Bold.otf',
                        #   regexp=r'[a-zA-Z\+]+|\b[\w\+]\b', 
                          color_func=pastel_color,
                          mask=mask,
                          random_state=42, 
                          max_font_size=35,  
                          ).generate(wordlist)
    
    plt.figure(figsize=(8,7))
    plt.imshow(wordcloud, interpolation='None')
    plt.axis('off')
    plt.savefig('heartwordcould.png',bbox_inches="tight")
    plt.show()

    

plot_word_cloud()
```
