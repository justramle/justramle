```python
import matplotlib.pyplot as plt 
from wordcloud import WordCloud
import random
```

### Generate Random pastel color for wordcloud

```python
def pastel_color(word, font_size, position, orientation, random_state=None, **kwargs):
    # range of HSL values for pink, pastel, and neutral tones
    h = random.randint(0, 360)  # Hue
    s = random.randint(50, 50)  # Saturation
    l = random.randint(60, 90)  # Lightness
    return f"hsl({h}, {s}%, {l}%)"
```

### Plotting word in wordcloud
- Example: https://stackoverflow.com/questions/65749888/how-to-generate-word-cloud-with-symbol
- Use regex pattern: any upper-lowecase  and + sign for C++ OR boundary with whole word include characters
    - regexp=r'[a-zA-Z\+]+|\b\w+\b'


```python
def plot_word_cloud():

    # Read file from txt
    with open ('worddoc.txt', 'r', encoding='UTF-8') as file:
        wordlist = ' '.join([line.strip() for line in file])
    
    # Specify phrases
    phrases = ['Hugging Face', 'Computer Science', 'Xin Chao','pairwise distance','open source',
               'Data Mining','Social Analysis','banh mi','Train model','Viet spicy noodle','Data modeling',
               'sticky rice','Bubble Tea','scikit Learn','cosine similarity','Jeju Island','carbon emission']

    # phrases = ['Hugging Face', 'Xin Chao', 'Computer Science','pairwise distance','open source',
    #            'Data Mining','Social Analysis','Train model','Data modeling',
    #            'scikit Learn','carbon emission']
    # Replace spaces with nospace for phrases
    for phrase in phrases:
        wordlist = wordlist.replace(phrase, phrase.replace(' ', ""))


    # Generate the word cloud
    wordcloud = WordCloud(width=800, height=400, background_color='white',
                          font_path='Montserrat-Bold.otf',
                          regexp=r'[a-zA-Z\+]+|\b[\w\+]\b', 
                          color_func=pastel_color
                          ).generate(wordlist)
    
    # plt.figure(figsize=(19.8, 4.95))
    plt.figure(figsize=(12,8))
    plt.imshow(wordcloud)
    plt.axis('off')
    plt.show()

plot_word_cloud()
```
